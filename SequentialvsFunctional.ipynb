```python
from tensorflow.keras.datasets import mnist

(X_train_full, y_train_full), (X_test, y_test) = mnist.load_data()
```

```python
X_valid, X_train = X_train_full[:5000] / 255.0, X_train_full[5000:] / 255.0
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
```

Sequential 1: Model with no hidden layer

```python
model = Sequential()
```

```python
model.add(Flatten(input_shape=[28, 28]))
#model.add(Dense(256, activation="relu"))
#model.add(Dense(128, activation="relu"))
model.add(Dense(10, activation="softmax"))
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model, to_file='model1_plot.png', show_shapes=True, show_layer_names=True)
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

```python
h = model.fit(X_train, y_train, epochs=10, validation_data=(X_valid, y_valid))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(h.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1) 
plt.show()
```

Functional: Complex deep model

```python
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Reshape
from tensorflow.keras.layers import Concatenate
from tensorflow.keras import Model
```

```python
input_ = Input(shape=[28, 28])
flatten = Flatten(input_shape=[28, 28])(input_)

#left
hidden10 = Dense(256, activation="relu")(flatten)
hidden11 = Dense(128, activation="relu")(hidden10)
hidden12 = Dense(64, activation="relu")(hidden11)
hidden13 = Dense(32, activation="relu")(hidden12)

#right
hidden14 = Dense(256, activation="sigmoid")(flatten)
hidden15 = Dense(128, activation="sigmoid")(hidden14)
hidden16 = Dense(32, activation="sigmoid")(hidden15)

concat = Concatenate()([hidden13,hidden16])

hidden17 = Dense(24, activation="relu")(concat)
hidden18 = Dense(16, activation="relu")(hidden17)

output = Dense(10, activation='softmax')(hidden17)

model4 = Model(inputs=[input_], outputs=[output] )

```

```python
model4.summary()
```

```python
plot_model(model4, to_file='model2_plot.png', show_shapes=True, show_layer_names=True)
```

```python
model4.compile(loss="sparse_categorical_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

```python
h4 = model4.fit(X_train, y_train, epochs=25, validation_data=(X_valid, y_valid))
```

```python
pd.DataFrame(h4.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1) 
plt.show()
```
